# Wrangle the data so that it follows consistent logic; for example, rename columns with illogical names and drop columns that donâ€™t add anything to your analysis.

```python
# Changing column names

customers.rename(columns={
    'First Name': 'first_name',
    'Surnam': 'last_name',
    'Gender': 'gender',
    'STATE': 'state',
    'Age': 'age',
    'date_joined': 'date_joined',
    'n_dependants': 'num_dependents',
    'fam_status': 'family_status'
}, inplace=True)
```

# Complete the fundamental data quality and consistency checks youâ€™ve learned throughout this Achievement; for example, check for and address missing values and duplicates, and convert any mixed-type data.

```python
# Checking datatypes
print(customers.dtypes)
```

```python
# Checking Nan
print(customers.isnull().sum())
```

```python
# Checking records
print(customers['gender'].unique())
print(customers['family_status'].unique())
```

```python
# changing date_joined datatype from object to datetime
customers['date_joined'] = pd.to_datetime(customers['date_joined'])
```

```python
# Renaming Nan in first_name to 'Unknown'
customers['first_name'] = customers['first_name'].fillna('Unknown')
```

```python
# Adjusting and alligning family_status records 
customers['family_status'] = customers['family_status'].replace({
    'living with parents and siblings': 'single',
    'divorced/widowed': 'divorced'
})
```
